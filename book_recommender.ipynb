# Popularity Based Recommendation System

## Ce système de recommandation basé sur la popularité sélectionne les 50 livres les plus populaires avec des notes moyennes provenant de livres ayant reçu plus de 150 critiques, offrant une liste soigneusement choisie de titres largement appréciés pour une découverte sans effort.



```python
# First merge ratings with books using ISBN column

ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
ratings_with_name
```

```python
# Effectuer un groupement sur ratings_with_name en utilisant le titre du livre pour compter les évaluations de chaque livre.

num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
# Effectuer un groupement sur ratings_with_name en utilisant le titre du livre et calculer la moyenne des évaluations.
avg_rating_df = ratings_with_name.groupby('Book-Title').agg({'Book-Rating': 'mean'}).reset_index()
```

```python
# Renaming Book-Rating as avg_rating

avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
# Ici, nous fusionnons num_rating_df avec avg_ratings en utilisant le titre du livre.

popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df.tail()
```

```python
# Ici, nous sélectionnons les livres évalués par plus de 150 utilisateurs, puis nous les trions en fonction de leurs évaluations moyennes par ordre décroissant.

# et enfin, nous affichons les 50 meilleurs livres, mettant en avant les titres les plus estimés.


popular_df = popular_df[popular_df['num_ratings']>=150].sort_values('avg_rating',ascending=False).head(50)
```

```python
# Fusionner le popular_df avec le dataframe des livres, car ce dernier contient toutes les informations nécessaires sur chaque livre requises pour l'affichage.

popular_df = popular_df.merge(books,on='Book-Title')
```

```python
# Dropping if there is any duplicates

popular_df = popular_df.drop_duplicates('Book-Title')
```

# Collaborative Filtering Based Recommendation System



## Filtrer les utilisateurs ayant évalué au moins 100 livres et sélectionner uniquement les livres ayant reçu plus de 50 évaluations, en créant un tableau croisé dynamique avec les livres comme index et les utilisateurs comme colonnes. Ce tableau croisé dynamique facilite le calcul de la similarité  pour recommander les  livres les plus similaires en fonction des requêtes des utilisateurs.


```python
# Filtrer uniquement les utilisateurs ayant évalué au moins 100 livres.

x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 100
```

```python
# Nous avons besoin de l'index de ces utilisateurs, donc nous extrayons l'index des utilisateurs filtrés (x).

padhe_likhe_users = x[x].index
```

```python
padhe_likhe_users
```

```python
# Ici, nous extrayons uniquement les informations des utilisateurs dont l'identifiant est dans padhe_likhe_users.

filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
filtered_rating
```

```python
# Comme nous souhaitons uniquement des livres célèbres, nous conservons ici uniquement ceux qui ont reçu plus de 50 évaluations.

y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
```

```python
# Pour obtenir les noms des livres, nous utilisons .index.

famous_books = y[y].index
```

```python
famous_books
```

```python
# Maintenant, nous extrayons uniquement les livres qui figuraient parmi les livres célèbres à partir des évaluations filtrées, afin d'obtenir notre dataframe final
# Nous avons retenu uniquement les utilisateurs ayant évalué au moins 100 livres et uniquement les livres ayant été évalués plus de 50 fois.

final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
# Ici, nous convertissons final_rating en tableau croisé dynamique pour le rendre adapté au filtrage collaboratif et à l'utilisation de la similarité.

# Dans l'index, nous prenons le titre du livre, et dans les colonnes, nous aurons tous les identifiants d'utilisateur. Les valeurs qui seront remplies dans ce tableau croisé dynamique seront les évaluations des livres.

pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt
```

```python
# Comme il y a de nombreuses valeurs NA, nous les remplissons par 0, car NA signifie qu'il n'y a pas d'évaluations, ce qui correspond à 0.

pt.fillna(0,inplace=True)
```

```python
pt
```

# Filtrage Collaboratif Article-Article (Item-Item)


```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
similarity_scores
```

```python
# Premièrement, nous calculons les scores de similarité des livres par rapport à tous les autres livres.

print(similarity_scores[0])
```

```python
similarity_scores.shape
```

```python
# Fonction principale qui calcule les détails des livres similaires en fonction du titre du livre passé en paramètre dans la fonction.
# Utiliser les scores de similarité calculés à l'aide de la similarité cosinus.


def recommend(book_name):
    # Récupération de l'index.
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]

    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)

    return data
```

```python
# Tester la fonction de recommandation, qui est notre fonction principale pour nous donner 4 recommandations en fonction de notre saisie.

recommend('Whispers')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
# Simuler une recommandation
book_name = '1984'
recommended_books = recommend(book_name)  # Obtenir des recommandations
recommended_titles = [book[0] for book in recommended_books]  # Récupérer les titres
# Calculer les métriques MAE et RMSE
actual_ratings = []
predicted_ratings = []
for title in recommended_titles:
    if title in pt.index:
        # Obtenir la note réelle (moyenne des notes)
        actual_rating = pt.loc[title].mean()
        actual_ratings.append(actual_rating)

        # Obtenir la note du livre de référence
        predicted_rating = pt.loc[book_name].mean()  # Note du livre de référence
        predicted_ratings.append(predicted_rating)
# Vérifier que nous avons des notes à comparer
if actual_ratings and predicted_ratings:
    mae = mean_absolute_error(actual_ratings, predicted_ratings)
    rmse = np.sqrt(mean_squared_error(actual_ratings, predicted_ratings))

    # Afficher les résultats
    print("Accuracy Metrics:")
    print(f"MAE: {mae:.4f}")
    print(f"RMSE: {rmse:.4f}")
else:
    print("Pas assez de données pour calculer les métriques.")
```

# Filtrage Collaboratif Utilisateur-Utilisateur (User-User)


```python
# Calculer la similarité entre utilisateurs
user_similarity_scores = cosine_similarity(pt.T)
user_similarity_scores
```

```python
print(user_similarity_scores[0])
```

```python
user_similarity_scores.shape
```

```python
def recommend_user(user_id):
    # Récupérer l'index de l'utilisateur
    index = np.where(pt.columns == user_id)[0][0]
    print(index)
    # Obtenir les utilisateurs similaires, triés par similarité
    similar_users = sorted(list(enumerate(user_similarity_scores[index])), key=lambda x: x[1], reverse=True)[1:6]
    print(similar_users)
    data = []
    for i in similar_users:
        # Récupérer les livres évalués par les utilisateurs similaires
        similar_user_id = pt.columns[i[0]]
        rated_books = pt[similar_user_id][pt[similar_user_id] > 5].index.tolist()[1:2]

        for book in rated_books:
            temp_df = books[books['Book-Title'] == book]
            item = []
            item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
            item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
            item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
            data.append(item)

    # Supprimer les doublons
    unique_data = [list(x) for x in set(tuple(x) for x in data)]

    return unique_data
```

# Recommendation  using Knn algorithm


```python
from scipy.sparse import csr_matrix
```

```python
# Convertir la matrice pivot en matrice creuse
sparse_matrix = csr_matrix(pt.values)
sparse_matrix
```

```python
from sklearn.neighbors import NearestNeighbors

# Appliquer KNN
knn = NearestNeighbors(n_neighbors=5, algorithm='brute', metric='cosine')
knn.fit(sparse_matrix)
```

```python
def recommendKnn(book_name):
    # Récupérer l'index du livre
    if book_name not in pt.index:
        return f"Le livre '{book_name}' n'existe pas dans le système."

    index = np.where(pt.index == book_name)[0][0]

    # Calculer les similarités
    distances, indices = knn.kneighbors(sparse_matrix[index], n_neighbors=5)  # Utiliser plus de voisins
    print(distances, indices)
    data = []
    for i in range(len(indices.flatten())):
        item = []
        temp_df = books[books['Book-Title'] == pt.index[indices.flatten()[i]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-L'].values))

        data.append(item)

    return data
```

# Recommendation System using ann


```python
import numpy as np
import pandas as pd
from sklearn.decomposition import TruncatedSVD
from sklearn.neural_network import MLPRegressor
```

```python
# 1. Créer des embeddings (en utilisant SVD)
svd = TruncatedSVD(n_components=50)  # Choisissez le nombre de dimensions
user_embeddings = svd.fit_transform(pt.T)  # Transposez pt pour que les utilisateurs soient en lignes
book_embeddings = svd.components_.T
```

```python
# 2. Préparer les données d'entraînement
data = []
for user_idx in range(len(user_embeddings)):
    for book_idx in range(len(book_embeddings)):
        rating = pt.iloc[book_idx, user_idx]  # Ajustez l'indexation pour la matrice transposée
        if rating > 0:  # Vérifiez que la note est positive
            data.append((user_embeddings[user_idx], book_embeddings[book_idx], rating))
```

```python
# 3. Entraînement du modèle ANN
X_train, y_train = zip(*[(np.concatenate((user, book)), rating) for user, book, rating in data])
X_train = np.array(X_train)
y_train = np.array(y_train)

model = MLPRegressor(hidden_layer_sizes=(50,), max_iter=1000)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# 4. Évaluation du modèle
# Prédire les évaluations pour tous les utilisateurs et livres
y_pred = []
y_true = []

for user_idx in range(len(user_embeddings)):
    for book_idx in range(len(book_embeddings)):
        actual_rating = pt.iloc[book_idx, user_idx]
        if actual_rating > 0:  # On ne prédit que pour les livres notés
            user_embedding = user_embeddings[user_idx]
            book_embedding = book_embeddings[book_idx]
            predicted_rating = model.predict([np.concatenate((user_embedding, book_embedding))])
            y_pred.append(predicted_rating[0])
            y_true.append(actual_rating)

# Convertir en tableaux numpy
y_pred = np.array(y_pred)
y_true = np.array(y_true)

# Calculer les métriques
mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
r2 = r2_score(y_true, y_pred)

# Afficher les résultats
print("Evaluation Metrics:")
print(f"MAE: {mae:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"R² Score: {r2:.4f}")
```

```python
# Récupérer les IDs de livres uniques
book_ids = np.unique(books['Book-Title'])

def recommend_with_embeddings(user_id, n_recommendations=5):
    user_id = int(user_id)
    user_embedding = user_embeddings[user_id]
    predictions = []

    # Prédire les évaluations pour tous les livres
    for book_idx in range(len(book_embeddings)):
        book_embedding = book_embeddings[book_idx]
        predicted_rating = model.predict([np.concatenate((user_embedding, book_embedding))])
        predictions.append((book_idx, predicted_rating[0]))  # Utiliser la première valeur de la prédiction

    # Obtenir les indices des livres les mieux notés
    recommended_indices = sorted(predictions, key=lambda x: x[1], reverse=True)[:n_recommendations]

    # Récupérer les informations sur les livres recommandés
    data = []
    for idx, rating in recommended_indices:
        item = []

        # Récupérer le titre du livre à partir de book_ids
        book_title = book_ids[idx]  # Utiliser l'index pour obtenir le titre
        print(book_title)

        # Extraire les informations du DataFrame des livres
        temp_df = books[books['Book-Title'] == book_title]

        if not temp_df.empty:  # Vérifier que le DataFrame n'est pas vide
            item.extend(list(temp_df['Book-Title'].values))
            item.extend(list(temp_df['Book-Author'].values))
            item.extend(list(temp_df['Image-URL-M'].values))

            data.append(item)

    return data
```

##   BERT Embedding Based Recommendation System:

```python
from transformers import BertTokenizer, BertModel
import torch
from sklearn.decomposition import TruncatedSVD
import numpy as np
from sklearn.neural_network import MLPRegressor
```

```python
# 1. Charger le modèle pré-entraîné et le tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
```

```python
# 2. Préparer les titres des livres pour obtenir des embeddings
book_titles = pt.index.tolist()  # Utiliser les titres comme index de pt
inputs = tokenizer(book_titles, padding=True, truncation=True, return_tensors="pt")
```

```python
# 3. Obtenir les embeddings des livres
with torch.no_grad():
    outputs = model(**inputs)
    book_embeddings = outputs.last_hidden_state.mean(dim=1)
```

```python
# 5. Intégrer ces embeddings dans votre système de recommandation
data = []

for user_idx in range(user_embeddings.shape[0]):  # Itérer sur les utilisateurs
    for book_idx in range(book_embeddings.shape[0]):  # Itérer sur les livres
        rating = pt.iloc[book_idx, user_idx]  # Récupérer la note
        if rating > 0:  # Vérifier que la note est positive
            data.append((user_embeddings[user_idx], book_embeddings[book_idx].numpy(), rating))
```

```python
# 6. Entraînement du modèle ANN
X_train, y_train = zip(*[(np.concatenate((user, book)), rating) for user, book, rating in data])
X_train = np.array(X_train)
y_train = np.array(y_train)

model = MLPRegressor(hidden_layer_sizes=(50,), max_iter=1000)
model.fit(X_train, y_train)
```

```python
# 7. Évaluation du modèle
y_pred = []
y_true = []

for user_idx in range(user_embeddings.shape[0]):
    for book_idx in range(book_embeddings.shape[0]):
        actual_rating = pt.iloc[book_idx, user_idx]
        if actual_rating > 0:
            user_embedding = user_embeddings[user_idx]
            book_embedding = book_embeddings[book_idx].numpy()
            predicted_rating = model.predict([np.concatenate((user_embedding, book_embedding))])
            y_pred.append(predicted_rating[0])
            y_true.append(actual_rating)

# Convertir en tableaux numpy
y_pred = np.array(y_pred)
y_true = np.array(y_true)

# Calculer les métriques
mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
r2 = r2_score(y_true, y_pred)

# Afficher les résultats
print("Evaluation Metrics:")
print(f"MAE: {mae:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"R² Score: {r2:.4f}")
```

```python
def recommend_with_embeddings(user_id, n_recommendations=5):
    user_id = int(user_id)
    user_embedding = user_embeddings[user_id]
    predictions = []

    for book_idx in range(len(book_embeddings)):
        book_embedding = book_embeddings[book_idx].numpy()
        predicted_rating = model.predict([np.concatenate((user_embedding, book_embedding))])
        predictions.append((book_idx, predicted_rating[0]))

    # Obtenir les indices des livres les mieux notés
    recommended_indices = sorted(predictions, key=lambda x: x[1], reverse=True)[:n_recommendations]

    data = []
    for idx, rating in recommended_indices:
        book_title = book_titles[idx]
        temp_df = books[books['Book-Title'] == book_title]

        if not temp_df.empty:
            # Récupérer les informations sans duplications
            unique_item = [
                book_title,
                temp_df['Book-Author'].iloc[0],  # Prendre le premier auteur
                temp_df['Image-URL-M'].iloc[0]   # Prendre la première image
            ]
            data.append(unique_item)

    return data
```
